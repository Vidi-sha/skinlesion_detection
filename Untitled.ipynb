```python
import pandas as pd
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from keras import optimizers 
from keras.layers import Dropout
from keras.optimizers import Adam
from keras.layers import Dense,Conv2D,Dropout,Flatten,MaxPooling2D
from keras.applications import MobileNetV2
```

```python
base_model=MobileNetV2(include_top=False, weights="imagenet", input_shape=(224,224,3))
```

```python
model = Sequential()
model.add(base_model)
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(7, activation='sigmoid'))
```

```python
import pandas as pd

# Load the metadata CSV file
data = pd.read_csv(r"C:\skin disease\HAM10000_metadata.csv")

# Create the 'HAM10000_images' column by combining 'image_id' with '.jpg'
data['HAM10000_images'] = data['image_id'] + '.jpg'

# Selecting the required columns into X
X = data[['HAM10000_images', 'dx', 'lesion_id']]
merged_data = pd.merge(X, data[['lesion_id', 'dx']], on='lesion_id', how='left')
# Now, 'X' contains 'dx', 'HAM10000_images', and 'lesion_id'
```

```python
X.sample(5)
```

```python
from sklearn.model_selection import train_test_split
Y=X.pop('dx').to_frame()

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.17, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.17, random_state=42)
```

```python
train=pd.concat([X_train, y_train],axis=1)
val=pd.concat([X_val, y_val],axis=1)
test=pd.concat([X_test,y_test],axis=1)

```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming you've already loaded and preprocessed your 'val' DataFrame

# Initialize LabelEncoder
encoder = LabelEncoder()

# Fit the encoder on training labels
encoder.fit(train['dx'])

# Transform both training and validation labels
train['label'] = encoder.transform(train['dx'])
val['label'] = encoder.transform(val['dx'])
```

```python
encoder= LabelEncoder()
encoder.fit(test['dx'])
name_as_indexes_test=encoder.transform(test['dx'])
test['label']=name_as_indexes_test

```

```python
from keras_preprocessing.image import ImageDataGenerator
train_generator = ImageDataGenerator(rescale =1./255,
                                     rotation_range=10,
                                     zoom_range = 0.1,
                                     width_shift_range=0.0, height_shift_range=0.00)
```

```python
train_data= train_generator.flow_from_dataframe(dataframe=train,x_col="HAM10000_images", y_col="dx",
                                                batch_size=32,directory="C:\skin disease\HAM10000_images_part_2",
                                                shuffle=True,class_mode="categorical",target_size=(224,224))
```

```python
test_generator=ImageDataGenerator(rescale = 1./255,rotation_range=10,
                                  zoom_range = 0.1,
                                  width_shift_range=0.0, height_shift_range=0.02)
```

```python
test_data= test_generator.flow_from_dataframe(dataframe=test,x_col="HAM10000_images", y_col="dx",
                                                directory="C:\skin disease\HAM10000_images_part_2",
                                                shuffle=False,batch_size=1,class_mode=None,target_size=(224,224))
```

```python
val_data= test_generator.flow_from_dataframe(dataframe=val,x_col="HAM10000_images", y_col="dx",
                                                directory="C:\skin disease\HAM10000_images_part_2",
                                                shuffle=False,batch_size=64,class_mode="categorical",target_size=(224,224))
```

```python
from keras.callbacks import ReduceLROnPlateau
learning_control=ReduceLROnPlateau(monitor='val_accuracy',patience=3,verbose=1,factor=.5,min_lr=0.0001)
```

```python
model.summary()
```

```python
from keras.optimizers import Adam,SGD
sgd = optimizers.SGD(lr=0.1, clipvalue = 0.5)
model.compile(optimizer=sgd,loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
histoy = model.fit_generator(generator=train_data,
                             steps_per_epoch=train_data.samples//train_data.batch_size,
                             validation_data=val_data,
                             verbose=1,
                             validation_steps=val_data.samples//val_data.batch_size,
                             epochs=40,callbacks=[learning_control])
```

```python
model.save('model12.h5')
```

```python
test_data.reset()
predictions = model.predict(test_data, steps=test_data.samples/test_data.batch_size,verbose=1)
y_predict= np.argmax(predictions,axis=1)
```
